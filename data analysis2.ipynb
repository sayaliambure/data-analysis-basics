```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = pd.read_csv('datasets/tips.csv')
tips.shape
```

```python
tips.head()
```

```python
titanic = pd.read_csv('datasets/titanic.csv')
titanic.shape
```

```python
flights = pd.read_csv('datasets/flights.csv')
flights.shape
```

```python
iris = pd.read_csv('datasets/iris.csv')
iris.shape
```

# bivarient and multivarient analysis

```python
# Scatterplot (Numerical - Numerical)
sns.scatterplot(tips['total_bill'], tips['tip'])
```

```python
# multivarient scatterplot
sns.scatterplot(tips['total_bill'], tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

```python
# Barplot (Numeric - Categorical)

sns.barplot(titanic['Pclass'], titanic['Age'])
```

```python
sns.barplot(titanic['Pclass'], titanic['Fare'], hue=titanic['Sex'])
```

```python
# Boxplot (Numeric - Categorical)
sns.boxplot(titanic['Sex'], titanic['Age'], hue=titanic['Survived'])
```

```python
# Distplot (Numeric - Categorical)

sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

```python
# Heatmap (Categorical - Categorical)

sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
print(titanic.groupby('Pclass').mean()['Survived']*100)
print(titanic.groupby('Sex').mean()['Survived']*100)
print(titanic.groupby('Embarked').mean()['Survived']*100)
```

```python
# clustermap (Categorical - Categorical)
# brings together the columns that perform similarly

sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
# pairplot - brings all scatter plots together, 
# used when we have a lot of numeric data 

sns.pairplot(iris, hue='species')
```

```python
# Lineplot (Numeric - Numeric)
# when we need a relation of data how it is with time, eg, stock analysis

sns.lineplot(flights['year'], flights['passengers'])
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(new['year'], new['passengers'])
```
